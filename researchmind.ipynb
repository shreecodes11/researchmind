```python
import arxiv
import pandas as pd
import numpy as np
import faiss
import torch

from sentence_transformers import SentenceTransformer
```

```python
import torch

print("Torch:", torch.__version__)

model = SentenceTransformer("all-MiniLM-L6-v2")

print("Sentence Transformer loaded!")
```

```python
print(arxiv)
```

```python
TOPIC = input("Enter Research Topic: ")

print("Topic:", TOPIC)
```

```python
search = arxiv.Search(
    query=TOPIC,
    max_results=20,
    sort_by=arxiv.SortCriterion.Relevance
)

client = arxiv.Client()

print("Search object created successfully!")
```

```python
papers = []

for result in client.results(search):

    paper = {
        "Title": result.title,
        "Authors": ", ".join(
            [author.name for author in result.authors]
        ),
        "Published": result.published.date(),
        "Abstract": result.summary,
        "PDF_URL": result.pdf_url
    }

    papers.append(paper)

print("Total Papers Retrieved:", len(papers))
```

```python
embedding_model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)

print("Embedding model loaded!")
```

```python
abstracts = df["Abstract"].tolist()

embeddings = embedding_model.encode(
    abstracts,
    show_progress_bar=True
)

print("Embedding Shape:", embeddings.shape)
```

```python
dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(
    np.array(embeddings).astype("float32")
)

print("Indexed Papers:", index.ntotal)
```

```python
faiss.write_index(
    index,
    "research_papers.index"
)

print("FAISS index saved!")
```

```python
query = input("Ask a research question: ")

print("Question:", query)
```

```python
query_embedding = embedding_model.encode(
    [query]
)

print(query_embedding.shape)
```

```python
k = 5

distances, indices = index.search(
    np.array(query_embedding).astype("float32"),
    k
)

print(indices)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
model_name = "google/flan-t5-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

print("LLM Loaded!")
```

```python
import nltk

from nltk.corpus import stopwords

from collections import Counter
```

```python
nltk.download("stopwords")
```

```python
stop_words = set(stopwords.words("english"))

words = all_text.lower().split()

clean_words = []

for word in words:

    word = word.strip(".,()[]{}:;!?\"'")

    if len(word) > 3 and word not in stop_words:

        clean_words.append(word)

print("Words cleaned!")
```

```python
word_counts = Counter(clean_words)

top_words = word_counts.most_common(20)

top_words
```

```python
print("Top Research Trends\n")

for word, count in top_words:

    print(f"{word}: {count}")
```

```python
research_phrases = [

    "large language model",

    "retrieval augmented generation",

    "rag",

    "llm ensemble",

    "agent",

    "multi agent",

    "reasoning",

    "fine tuning",

    "knowledge graph",

    "hallucination",

    "benchmark",

    "inference",

    "alignment",

    "tool learning",

    "prompt engineering",

    "recommendation system",

    "graph neural network"
]
```

```python
phrase_counts = {}

text = all_text.lower()

for phrase in research_phrases:

    phrase_counts[phrase] = text.count(
        phrase.lower()
    )
```

```python
sorted_trends = sorted(
    phrase_counts.items(),
    key=lambda x: x[1],
    reverse=True
)
```

```python
print("Research Trend Analysis\n")

for phrase, count in sorted_trends:

    if count > 0:

        print(
            f"{phrase}: {count}"
        )
```
